## Gradient Descent

$\large J(\theta)$ is called **cost function** or **target function**. It is the function that we seek to minimize.<br>

Find $\hat{\mathbf{\theta}}$ that minimizes $J(\theta)$ : $\underset{\theta}{\arg\max}$$J(\theta)$

## Excercise 1

Find the minimum value of $\large f_{(x, y)} = (x - 2)^{2} + (y - 2)^{2} + 4$ using **Gradient Descent**

### 1.1 Find the gradient

$$\large \frac{d}{dx}f_{(x, y)} = 2(x - 2)$$<br>
$$\large \frac{d}{dy}f_{(x, y)} = 2(y - 2)$$

```python
#create a gradient function
def get_gradient(v):
    """returns gradient at a given point"""
    return [2 * (v_i - 2) for v_i in v]
```

```python
#for example
get_gradient([1,2])
```

### 1.2 Find the steps

$$\large x_{t+1} = x_{t} - \rho \frac{d}{dx}f_{(x,y)} = x_{t} - \rho \times 2(x - 2)$$<br><br>
$$\large y_{t+1} = y_{t} - \rho \frac{d}{dy}f_{(x,y)} = y_{t} - \rho \times 2(y - 2)$$

```python
def step(v, gradient, step_size):
    """moves from the current position with respect to gradient and step_size values"""
    return [v_i - step_size * direction_i 
            for v_i, direction_i in zip(v, gradient)]
```

### 1.3 Start from random position

```python
import numpy as np

random_generator = np.random.RandomState(42)

#create a starting point x_0, y_0
v = [random_generator.randint(-10, 10) for _ in range(2)]
v
```

### 1.4 Set alpha

```python
alpha = 0.01
```

### 1.5 Set threshold to stop algorithm

$$\large dist_{(v_t, v_{t+1})} < tolerance$$ <br>
$$\large v = (x, y)$$

```python
import math
def distance(v, w):
    return math.sqrt(sum([(v_i - w_i)**2 for v_i, w_i in zip(v, w)]))
```

```python
#recall our starting point v
v
```

### 1.6 Gradient Descent Algorithm
<br>
1. We set a starting point $v = [-4, 9]$<br><br>
2. gradient = get_gradient : This will calculate gradient at the current position<br><br>
3. next_v = step(v, gradient, alpha) : This will move $v$ with respect to gradient and alpha values<br><br>
4. plot_point.append(next_v) : This will store the new v<br><br>
5. distance(next_v, v) : This will calculate the distance between next_v and the current v. If the distance is less than 0.00001, then we stop the loop

```python
cnt = 0
tol = 0.00001 #set tolerance
gradients = []  #in order to check changes of gradient values
plot_points = []  #for plotting purpose

while True:
    cnt += 1 #update count
    gradient = get_gradient(v)  #gradient at current point
    next_v = step(v, gradient, alpha)
    if not cnt % 15: #15%15 is 0, which is False. if not false, proceed the block
        plot_points.append(next_v)
        print(next_v)  #for debugging
    if distance(next_v, v) < tol:
        break
    gradients.append(next_v) #store next values of v into the gradient list
    v = next_v
```

```python
#all next values
gradients
```

### 1.7 Contour plot results

```python
gradients = np.array(gradients)

x = gradients[:, 0]
y = gradients[:, 1]
```

```python
#use meshgrid to plot contours
X, Y = np.meshgrid(x, y)
Z = (X-2) ** 2 + (Y-2) ** 2
```

```python
#bring plot_points to check the interim values that we stored in the algorithm
plot_points = np.array(plot_points)

px = plot_points[:, 0]
py = plot_points[:, 1]
```

```python
import matplotlib.pyplot as plt
```

```python
#plot contour
contours = plt.contourf(x, y, Z, 20, cmap='RdGy')
plt.plot(px, py, "b-o")
plt.colorbar();
```

## Excercise 2

Find the minimum value of $\large f_{(x, y)} = (1 - x)^{2} + 100(y - x^{2})^{2}$ using **Gradient Descent**

### 2.1 Find the gradient

$$\large \frac{d}{dx}f_{(x, y)} = 400x^{3} - 400xy + 2x -2$$<br>
$$\large \frac{d}{dy}f_{(x, y)} = 200y - 200x^{2}$$

```python
#create a gradient function
def get_gradient(v):
    """returns gradient at a given point"""
    return [400 * v[0] ** 3 - 400 * v[0] * v[1] + 2 * v[0] - 2, 200 * v[1] - 200 * v[0] ** 2]
```

```python
#for example
get_gradient([1,2])
```

### 2.2 Find the steps

$$\large x_{t+1} = x_{t} - \rho \frac{d}{dx}f_{(x,y)} = x_{t} - \rho \times 400x^{3} - (400xy + 2x - 2)$$<br><br>
$$\large y_{t+1} = y_{t} - \rho \frac{d}{dy}f_{(x,y)} = y_{t} - \rho \times (200y - 200x^{2})$$

```python
def step(v, gradient, step_size):
    """moves from the current position with respect to gradient and step_size values"""
    return [v_i - step_size * direction_i 
            for v_i, direction_i in zip(v, gradient)]
```

### 2.3 Start from random point

```python
import numpy as np

random_generator = np.random.RandomState(42)

#create a starting point x_0, y_0
v = [random_generator.randint(-10, 10) for _ in range(2)]
v
```

### 2.4 Set alpha

```python
alpha = 0.0001
```

### 2.5 Set threshold to stop algorithm

$$\large dist_{(v_t, v_{t+1})} < tolerance$$ <br>
$$\large v = (x, y)$$

```python
import math
def distance(v, w):
    return math.sqrt(sum([(v_i - w_i)**2 for v_i, w_i in zip(v, w)]))
```

```python
cnt = 0  
tol = 0.000001

gradients = []  # change of gradient
plot_points = []

while True:
    cnt += 1
    gradient = get_gradient(v)  #gradient
    next_v = step(v, gradient, alpha)
    if not cnt % 2000:
        plot_points.append(next_v)
        gradients.append(next_v)
        print(next_v)  # Debugging
    if distance(next_v, v) < tol:
        break
    
    v = next_v
```

### 2.6 Contour plot results

```python
gradients = np.array(gradients)

x = gradients[:, 0]
y = gradients[:, 1]
```

```python
#to meshgrid
X, Y = np.meshgrid(x, y)
Z = (1-X) ** 2 + 100*(Y-X ** 2) ** 2
```

```python
plot_points = np.array(plot_points)

px = plot_points[:, 0]
py = plot_points[:, 1]
```

```python
contours = plt.contourf(x, y, Z, 20, cmap='RdGy')
plt.plot(px, py, "b-o")
plt.colorbar();
```
